```python
def quality_of_tip(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "Stingy"
```

```python
quality_of_tip(16.99,1.01)
```

```python
df["Quality"] = df[["total_bill","tip"]].apply(lambda df: quality_of_tip(df["total_bill"],df["tip"]),axis=1)
```

```python
df.head()
```

```python
df["Quality"] = np.vectorize(quality_of_tip)(df["total_bill"],df["tip"])
```
