```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(4,4),dpi=200)

axes[0,0].plot(x,y)
axes[0,0].set_title("Hello")
axes[0,0].set_xlabel("X Stuff")
axes[0,0].set_ylabel("Y Stuff")

axes[1,1].plot(a,b)

fig.subplots_adjust(wspace=.5,hspace=.5)

fig.suptitle("My 4 plots",fontsize=15)

# plt.tight_layout()

fig.savefig("All 4 Subplots.png",bbox_inches="tight")
```
