```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
fig = plt.figure()

# LARGE AXES 
axes_1 = fig.add_axes([0,0,1,1])
axes_1.plot(a,b)

axes_1.set_xlim(0,9)
axes_1.set_ylim(0,9000)
axes_1.set_xlabel("A")
axes_1.set_ylabel("B")
axes_1.set_title("Power of 4")


# SMALL AXES
axes_2 = fig.add_axes([1,1,0.5,0.5])
axes_2.plot(a,b)

axes_2.set_xlim(1,2)
axes_2.set_ylim(0,50)
axes_2.set_xlabel("A")
axes_2.set_ylabel("B")
axes_2.set_title("Zoomed In")

```
